Document Clustering With Python

Dataset is movies data set web scrapped from net into two files - titles.txt and synopses.txt 

```python
#import libraries
import numpy as np
import pandas as pd
import nltk
from nltk.stem.snowball import SnowballStemmer
import json
import re
import string
from sklearn.feature_extraction.text import TfidfVectorizer
nltk.download('stopwords')
nltk.download('punkt')
```

```python
# reading the data from 2 files = titles.txt & synopses.txt
with open('titles.txt', 'r') as filehandle:
    titles = json.load(filehandle)
    
with open('synopses.txt', 'r') as filehandle:
    synopses = json.load(filehandle)
```

```python
#stopwords, stemming, & tokenizing

stops = nltk.corpus.stopwords.words('english')+list(string.punctuation)

stemmer = SnowballStemmer('english')
```

```python
def myTokenizer(text):
    temp_tokens = [word for word in nltk.word_tokenize(text)]
    tokens = []
    for token in temp_tokens:
        if re.search('[a-zA-Z]', token):
            tokens.append(token)
    ret_tokens = [stemmer.stem(t) for t in tokens]
    return ret_tokens
```

```python
synopses[0]
```

```python
syn0 = synopses[0]
myTokenizer(syn0)
```

```python
tfidf_vectorizer = TfidfVectorizer(max_df = 0.8, max_features = 200000,
                                  min_df = 0.05, stop_words = stops,
                                  use_idf = True, tokenizer = myTokenizer,
                                  ngram_range= (1,3))
tfidf_matrix = tfidf_vectorizer.fit_transform(synopses)
terms = tfidf_vectorizer.get_feature_names()
```

```python
tfidf_matrix.shape
```

```python
terms
```

```python
#Hierarchical document clustering

from sklearn.metrics.pairwise import cosine_similarity
dist = 1 - cosine_similarity(tfidf_matrix)
```

```python
import matplotlib.pyplot as plt

from scipy.cluster.hierarchy import ward, dendrogram
#define the linkage_matrix using ward clustering pre-computed distances
linkage_matrix = ward(dist)

fig, ax = plt.subplots(figsize=(15,20))
ax = dendrogram(linkage_matrix, orientation = 'right', labels = titles);

plt.tick_params(\
               axis= 'x',
               which = 'both',
               bottom = False,
               top = False,
               labelbottom = False)

plt.tight_layout()

plt.savefig('ward_cluster.png', dpi = 200)
```
